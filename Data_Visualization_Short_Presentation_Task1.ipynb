```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
fruits = ['UK','USA','Japan','India']
counts = [40, 100, 30, 55]
bar_labels = ['UK','USA','Japan','India']
bar_colors = ['tab:red','tab:blue','tab:red','tab:orange']

ax.bar(fruits, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('Number of Viewers(K)')
ax.set_title('Netflix Viewers in different Countries')
ax.legend(title='Countries')

plt.show()
```

```python
##Bar Label
import numpy as np
species = ('USA', 'UK', 'India')
Netflix = {
    'Movies': np.array([73, 34, 61]),
    'TVShows': np.array([73, 34, 58]),
}
width = 0.6  # the width of the bars: can also be len(x) sequence


fig, ax = plt.subplots()
bottom = np.zeros(3)

for show, Net in Netflix.items():
    p = ax.bar(species, Net, width, label=show, bottom=bottom)
    bottom += Net

    ax.bar_label(p, label_type='center')

ax.set_title('Number of Netflix Viewers by TV Shows and Movies')
ax.set_ylabel('Number of Viewers(K)')
ax.set_xlabel('Countries')
ax.legend()

plt.show()
```
